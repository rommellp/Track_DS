```python
import streamlit as st
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import os
import random
import bar_chart_race as bcr
```

```python
os.chdir(r"C:\Users\5luca\Documents\Python\Projects\Track_DS\Streamlit_folder\merged_df")
df = pd.read_csv('Runner_Comp_400m.csv')
df.info()
```

```python
def pace_array(x):
    distance = 400
    pace = distance / x
    array = np.arange(0,distance,pace)
    array = np.append(array,distance)
    return array
```

```python
def adding_NAN(x):
    y = max_len - len(x)

    nan_values = np.full((y), np.nan)
    new_arr = np.concatenate((x, nan_values))

    return new_arr
```

```python
competitors = df.sample(7)
competitors
```

```python
IDComp = competitors.drop(columns=['Gender', 'Grade Level','Season','400 Meters'],axis = 1)
IDComp = IDComp.T
IDComp.columns=IDComp.iloc[0]
IDComp = IDComp.drop(index=['ID Number'])

time0 =competitors.iloc[0,4]
time1 =competitors.iloc[1,4]
time2 =competitors.iloc[2,4]
time3 =competitors.iloc[3,4]
time4 =competitors.iloc[4,4]
time5 =competitors.iloc[5,4]
time6 =competitors.iloc[6,4]

pace0list = pace_array(time0)
pace1list = pace_array(time1)
pace2list = pace_array(time2)
pace3list = pace_array(time3)
pace4list = pace_array(time4)
pace5list = pace_array(time5)
pace6list = pace_array(time6)

array_com = [pace0list,pace1list,pace2list,pace3list,pace4list,pace5list,pace6list]
max_len = max(len(subarr) for subarr in array_com)

arr0 = adding_NAN(pace0list)
arr1 = adding_NAN(pace1list)
arr2 = adding_NAN(pace2list)
arr3 = adding_NAN(pace3list)
arr4 = adding_NAN(pace4list)
arr5 = adding_NAN(pace5list)
arr6 = adding_NAN(pace6list)

arr0 = np.vstack(arr0)
arr1 = np.vstack(arr1)
arr2 = np.vstack(arr2)
arr3 = np.vstack(arr3)
arr4 = np.vstack(arr4)
arr5 = np.vstack(arr5)
arr6 = np.vstack(arr6)

IDComp.iloc[:,[0]] = arr0
IDComp.iloc[:,[1]] = arr1
IDComp.iloc[:,[2]] = arr2
IDComp.iloc[:,[3]] = arr3
IDComp.iloc[:,[4]] = arr4
IDComp.iloc[:,[5]] = arr5
IDComp.iloc[:,[6]] = arr6

df = pd.DataFrame(IDComp)

df = df.rename(columns={df.columns[0]: 'Runner1',
                        df.columns[1]: 'Runner2', 
                        df.columns[2]: 'Runner3',
                        df.columns[3]: 'Runner4', 
                        df.columns[4]: 'Runner5',
                        df.columns[5]: 'Runner6',
                        df.columns[6]: 'Runner7',})

df = df.round(decimals = 2)
with pd.option_context('display.max_rows', None,):
    print(df)
```
